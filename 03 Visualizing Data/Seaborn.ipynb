```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(sns)
```

```python
df = pd.read_csv('/home/martin/datasets/flights.csv')
df.head()
```

```python
sns.relplot(x="arr_delay", y="carrier", data=df)
```

```python
airports_df = df[(df['origin'] == 'JFK')
        | (df['origin'] == 'LGA')
        | (df['origin'] == 'EWR')
       ]
sns.catplot(x='origin', y='sched_dep_time', kind='violin', data=airports_df)
```

```python
hour_delay = df[df['arr_delay'] > 60]
sns.relplot(x="sched_dep_time", y="arr_delay", hue='carrier', data=hour_delay)
```

```python
sns.jointplot(x="sched_dep_time", y="arr_delay", data=hour_delay)
```

```python
sns.relplot(x="dep_delay", y="arr_delay", data=df[:150], kind="line")
```
